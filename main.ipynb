```python
import matplotlib.pyplot as plt
```

```python
reaction_counts = cleaneddf['Reaction Type'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(reaction_counts, labels=reaction_counts.index, autopct='%1.1f%%', startangle=90)

plt.title('Distribution of Reaction Types')
plt.axis('equal')

# Display the chart
plt.show()
```

```python
reaction_counts = cleaneddf['Content Type'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(reaction_counts, labels=reaction_counts.index, autopct='%1.1f%%', startangle=90)

plt.title('Distribution of Content Types')
plt.axis('equal')

# Display the chart
plt.show()
```

```python
num_unique_categories = cleaneddf['Category'].str.lower().nunique()

cleaneddf['Category'] = cleaneddf['Category'].str.lower()

category_counts = cleaneddf['Category'].value_counts()
plt.bar(category_counts.index, category_counts)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title(f'Number of Unique Categories: {num_unique_categories}')
plt.xticks(rotation=90)
plt.show()
```

```python
cleaneddf['Datetime'] = pd.to_datetime(cleaneddf['Datetime'])

cleaneddf.set_index('Datetime', inplace=True)

monthly_counts = cleaneddf['Content ID'].resample('M').count()

monthly_counts.plot(kind='line', figsize=(10, 6))

plt.xlabel('Month')
plt.ylabel('Number of Posts')
plt.title('Number of Posts by Month')

plt.show()
```

```python
filtered_df = cleaneddf[cleaneddf['Category'].isin(top5df['Category'])]

sentiment_counts = filtered_df.groupby(['Category', 'Sentiment']).size().unstack()

most_popular_category = top5df.iloc[0]['Category']

sentiments = sentiment_counts.loc[most_popular_category]

fig, ax = plt.subplots()
ax.pie(sentiments, labels=sentiments.index, autopct='%1.1f%%')

ax.set_title('Sentiment Distribution the Most Popular Category')

plt.show()
```

```python
sentiment_counts = cleaneddf.groupby(['Category', 'Sentiment']).size().unstack()

category = cleaneddf.iloc[0]['Category']

sentiments = sentiment_counts.loc[category]

fig, ax = plt.subplots()
ax.pie(sentiments, labels=sentiments.index, autopct='%1.1f%%')

ax.set_title('Sentiment Distribution for Individual Category')

plt.show()
```
